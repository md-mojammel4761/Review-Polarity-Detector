```python
import pandas as pd
import re 
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('../datas/raw_data/IMDB_Dataset.csv')
```

```python
df.head()
```

```python
def clean_text(text):
    text = text.lower()                          # Convert to lowercase
    text = re.sub(r'<.*?>', '', text)            # Remove HTML tags
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub(r'([!?.])\1+', r'\1', text)    # Reduce repeated punctuation
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)   # Limit repeated characters
    text = re.sub(r'\d+/\d+|\d+', '', text)      # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)          # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()     # Normalize spaces
    
    word = text.split()
    stop_word= set(stopwords.words("english"))
    words = [w for w in word if w not in stop_word]

    WNL  = WordNetLemmatizer()
    words = [WNL.lemmatize(w) for w in words]

    return " ".join(words)
```

```python
df["clean_text"] = df["review"].apply(clean_text)
```
