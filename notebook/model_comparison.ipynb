```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

import pickle
```

```python
model = pickle.load(open("../models/logestic_reg.pkl","rb"))
```

```python
tfidf_vec = pickle.load(open("../models/tfid_vec.plk","rb"))
```

```python
lb_encoder = pickle.load(open("../models/lb.pkl","rb"))
```

```python
df = pd.read_csv("../datas/clean_data/clean_dataset.csv")
```

```python
df.head()
```

```python
x = df["clean_text"]
y = lb_encoder.transform(df["sentiment"])
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.2,random_state=34,stratify=y)
```

```python
xtest_tfidf = tfidf_vec.transform(xtest)
```

```python
pred_y = model.predict(xtest_tfidf)
```

```python
print(f'accuracy is score is: {accuracy_score(ytest,pred_y)}')

cm = confusion_matrix(ytest,pred_y)
print(f'confusion matrix is: {cm}')
```

```python
print(f'Classification Report:\n{classification_report(ytest, pred_y, target_names=lb_encoder.classes_)}')
```

```python
false_pos_idx = (pred_y == 1) & (ytest == 0)
print(f'Sample False Positive Reviews:\n{xtest[false_pos_idx].head(5)}')
```

```python
false_neg_idx = (pred_y == 0) & (ytest == 1)
print(f'Sample False Negative Reviews:\n{xtest[false_neg_idx].head(5)}')
```
